```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reaction=pd.read_csv("Task 3_Final Content Data set.csv")
reaction=reaction.drop(columns="Unnamed: 0")
category=reaction.groupby('Category')[['Score']].sum()
category=category.sort_values(by='Score',ascending=False)
top5category=category.head().reset_index()
top5category
```

```python
top5category.plot(kind='bar',
                  x='Category')

plt.title('Top 5 categories by aggregate popularity score')
plt.ylabel('Aggregate score')
plt.xlabel('Categories')
plt.show()
```

```python
explode=(0.1,0,0,0,0)
plt.pie(top5category['Score'], labels=top5category['Category'], autopct='%1.1f%%', startangle=140,explode=explode,shadow=True)
plt.title('Popularity percent share by top 5 categories')

plt.xlabel('Categories')
plt.show()
```
